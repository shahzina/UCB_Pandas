```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the books.csv file path
file = "Resources/books.csv"
# Import the books.csv file as a DataFrame
file_df = pd.read_csv(file, encoding="ISO-8859-1")
file_df.head()
file_df.columns
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
# Only keep: ""isbn"", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"
clean_file = file_df.drop(columns =['id','book_id','best_book_id', 'work_id','books_count', 'isbn13',
                                    'ratings_count','work_ratings_count','language_code','title',
                                    'work_text_reviews_count','image_url','small_image_url'])
clean_file.head()
```

```python
# Rename the headers to be more explanatory

renamed_file = clean_file.rename(columns = {"isbn":"ISBN code", "authors": "Authors",
                                            "original_publication_year": "Year of publication",
                                            "original_title": "title", "ratings_1":"ratings 1","ratings_2": "ratings 2",
                                            "ratings_3": "ratings 3", "ratings_4": "ratings 4",
                                            "ratings_5": "ratings 5"})
renamed_file.head()
```

```python
# Push the remade DataFrame to a new CSV file
file_df.to_csv("Resources/GoodReadsOutput.csv", index=False, header=True)
```
